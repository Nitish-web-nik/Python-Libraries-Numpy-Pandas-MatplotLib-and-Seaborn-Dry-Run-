# ----Matplotlib----

```python
#Matplotlib is a python library used for data visualization
#You can create bar-plots, scatter-plots, histograms and a lot more with matplotlib
```

```python
#Line Plot 
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = 3 * x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
#Adding Title and Labels
```

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-axis", c = "red")
plt.ylabel("y-axis", color = 'green')
plt.show()
```

```python
#Changing line Aesthetics
plt.plot(x,y,color = 'g', linestyle = ':', linewidth =4 )
plt.show()
```

```python
#Adding two lines in the same plot
```

```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
```

```python
plt.plot(x,y1, color = 'g', linestyle = ':', linewidth = 4)
plt.plot(x,y2, color = 'y', linestyle = '-', linewidth = 3)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
#Adding sub-plots in the line plot
x = np.arange(1,20)
y1 = 3 * x
y2 = 4 * x

plt.subplot(1,2,1) # means make 1 rows and 2 columns type subplots and last 1 indicates this is the first part
plt.plot(x, y1, color = 'g', linestyle = ':', linewidth = 3)

plt.subplot(1,2,2)
plt.plot(x, y2, color = 'y', linestyle = '-', linewidth = 4)


plt.show()
```

```python
#Above one with column wise
x = np.arange(1,20)
y1 = 3 * x
y2 = 4 * x

plt.subplot(2,1,1) # means make 1 rows and 2 columns type subplots and last 1 indicates this is the first part
plt.plot(x, y1, color = 'g', linestyle = ':', linewidth = 3)

plt.subplot(2,1,2)
plt.plot(x, y2, color = 'y', linestyle = '-', linewidth = 4)


plt.show()
```

```python
#loading the required libraries
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
ipl = pd.read_csv('matches.csv')
```

```python
ipl.head()
```

```python
ipl.shape
```

```python
#Getting the freq of most MOM awards
ipl['player_of_match'].value_counts()
```

```python
ipl['player_of_match'].value_counts()[0:10] #top 10 most times MOM Awards
```

```python
#To get only the names of the players
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
plt.figure(figsize = (8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]), color = 'g')
plt.show()
```

```python
#Getting the frequency of result column
ipl['result'].value_counts()
```

```python
#Finding out the number of toss wins w.r.t each team
ipl['toss_winner'].value_counts()
```

```python
#Extracting the records where a team won batting first
batting_first = ipl[ipl['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
#Making a histogram
plt.figure(figsize=(5,7))
plt.hist(batting_first['win_by_runs'])
plt.title('Distribution of Runs')
plt.xlabel('Runs')
plt.show()
```

```python
#Finding out the number of wins w.r.t each team after batting first
batting_first['winner'].value_counts()
```

```python
#Making a bar-plot for top 3 teams with most wins after batting first
plt.figure(figsize=(6,6))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]), color = ['blue', 'yellow', 'red'])
plt.show()
```

```python
#Making a pie-chart
plt.figure(figsize= (7,7))
plt.pie(list(batting_first['winner'].value_counts()), labels = list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#Extracting those records where a team has won after batting second
```

```python
batting_second = ipl[ipl['win_by_wickets']!=0]
```

```python
#looking at the head
batting_second.head()
```

```python
#Making a histogram for frequency of wins w.r.t number of wickets
plt.figure(figsize=(5,5))
plt.hist(batting_second['win_by_wickets'],bins = 30)
plt.show()
```

```python
#Finding out the frequency of number of wins w.r.t each tie after batting second 
batting_second['winner'].value_counts()
```

```python
#Making a bar plot for top-3 teams with most wins after batting second 
plt.figure(figsize = (6.5,6.5))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color = ['purple', 'blue','red'])
plt.show()
```

```python
#Making a pie-chart
plt.figure(figsize= (6.5,6.5))
plt.pie(list(batting_second['winner'].value_counts()), labels = list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
